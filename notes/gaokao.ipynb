```python
from PyPDF2 import PdfReader, PdfWriter

# Get current directory
import os
import builtins

cd = os.getcwd()
fn = os.path.join(cd, '../books/gaokao/北京市2024年普通高等学校招生专业目录普通类（本科）.pdf')
print(fn)

inputpdf = PdfReader(open(fn, "rb"))

output = PdfWriter()
output.add_page(inputpdf.pages[31])
with builtins.open('../books/tsinghua.pdf', 'wb') as f:
    output.write(f)
```
